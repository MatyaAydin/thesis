```python
import numpy as np
import matplotlib.pyplot as plt
```

```python


def plot_hadamard_ellipsoids(X, D):
    theta = np.linspace(0, 2*np.pi, 100)
    sphere_points = np.array([np.cos(theta), np.sin(theta)])

    ellipsoid_x = X @ sphere_points

    X_hadamard = X / D
    ellipsoid_xd = X_hadamard @ sphere_points
    
    
    plt.plot(ellipsoid_x[0, :], ellipsoid_x[1, :], label='X')
    plt.plot(ellipsoid_xd[0, :], ellipsoid_xd[1, :], label='X * D', color='red', linestyle='--')

    plt.legend()
    plt.axis('equal')
    plt.show()

X = np.array([[2, 1],
              [1, 2]])
D = np.array([[1.5, 0.5],
              [0.5, 1.5]])
plot_hadamard_ellipsoids(X, D)
```
